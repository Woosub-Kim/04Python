```python
print(df.groupby('AM or PM').Job.count())
print(df.groupby('Job').Language.count())
print(df.groupby('Language').Job.count())
```

```python
# 회계관련 데이터 concat하기
cca = df[df.Job=='Accountant, chartered certified']
cma = df[df.Job=='Accountant, chartered management']
ca = df[df.Job=='Accountant, chartered']

pd.concat([cca,cma,ca])
```

```python
# 평균 구매가격 최고 구매가격 최소 구매가격
print('$',df['Purchase Price'].mean()) #평균가
print('$',df['Purchase Price'].max()) #최고가
print('$',df['Purchase Price'].min()) #최저가
print('$',df[df['Purchase Price']!=0]['Purchase Price'].min()) #0이아닌 최저가
```

```python
# 영어권 사용자의 평균구매금액 
print(df.Language.unique()) #언어의 종류
print('$',df[df.Language=='en']['Purchase Price'].mean()) #영어(en)의 평균가
```

```python
# 언어별 구매금액의 평균
print(df.groupby('Language')['Purchase Price'].mean())
```

```python
# Job이 'Lawyer'인 사람의 숫자는
#print(df.Job.unique()) #직업 종류추출(너무많음)
#1. 'Lawyer'을 찾아서 카운트
print(df[df.Job=='Lawyer'].Lot.count())
print(sum(df.Job=='Lawyer'))

# 대소문자까지 고려한 카운트
print(sum(df.Job.apply(lambda x:x.lower()) == 'lawyer'))
# Job 문자열에 'Lawyer'포함된 경우까지 카운트(for문, in 키워드)
count = 0
for j in df.Job:
    if 'lawyer' in j.lower():
        count += 1
print(count)
print(len([j for j in df.Job if 'lawyer' in j.lower()]))
```

```python
# Job별로 인원 출력
print(df.groupby('Job').Lot.count())
```

```python
# AM과 PM 각각 구매한 건수
# 방법1 group by
print(df.groupby('AM or PM').Lot.count())

#방법2 for문
AM = 0
PM = 0

for t in df['AM or PM']:
    if t=='AM':
        AM += 1
    else:
        PM += 1
print(AM, PM)
# 방법3 Boolean Seriese만들어 sum()적용
print('AM',sum(df['AM or PM']=='AM'))
print('PM',sum(df['AM or PM']=='PM'))
# 방법4 치트키
print(df['AM or PM'].value_counts())
```

```python
# Lot이 '90 WT'인 곳에서 구매한 구매금액
print(df[df.Lot=='90 WT']['Purchase Price'])
print(df.groupby('Lot')['Purchase Price'].mean()['90 WT'])
```

```python
# 카드번호가 4926535242672853'인 사람의 이메일 주소
print(df[df['Credit Card']==4926535242672853].Email)
```

```python
# 카드회사가 'American Express'이고 구매금액이 95$ 이상인 고객수
print(sum( (df['CC Provider']=='American Express') & (df['Purchase Price']>=95) ))
print(df[(df['CC Provider']=='American Express')&(df['Purchase Price']>=95)].count().Lot)
print(sum( df[df['CC Provider']=='American Express']['Purchase Price'] >= 95))
```

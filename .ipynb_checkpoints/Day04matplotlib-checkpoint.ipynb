```python
import matplotlib.pyplot as plt
```

```python
plt.figure() #주피터 노트북의 기능: 자동생성된다.
x = [1,2,3,4,5]
y = [6,7,8,9,10]
z = [i**2 for i in x]
plt.plot(x,y)
plt.figure()
plt.plot(x,z)
```

```python
import numpy as np
x1 = range(50)
y1 = [np.cos(val) for val in x1]
x2 = range(50)
y2 = [np.sin(val) for val in x2]
plt.figure()
plt.plot(x1,y1)
plt.plot(x2,y2)
```

```python
plt.subplot(3,1,1)
plt.plot(x1,y1, color = 'red', linestyle='dotted')
plt.subplot(3,1,3)
plt.plot(x2,y2, color='blue', linestyle='dashed')
```

```python
plt.figure(figsize=(4,6))
x1 = range(100)
y1 = [val for val in x1]
x2 = range(100)
y2 = [val + 50 for val in x2]
plt.plot(x1,y1, c='#a52a2a', ls='-.', label='Line 1')
plt.plot(x2,y2, c='#0000ff', ls=':', label = 'Line 2')

plt.annotate('Line 2', xytext=(30,110), xy=(50,100),arrowprops=dict(facecolor='yellow', shrink=0.1))
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend(fancybox = True, shadow=True, loc='upper right')
plt.savefig('matpolib.jpg')
```

# Scatter

```python
x = list(range(1,6))
y = list(range(6, 11))
plt.scatter(x,y)
```

```python
x = np.random.rand(100)
y = x + np.random.randn(100)/10
plt.scatter(x,y)
```

```python
data1 = np.random.rand(512,2)
data2 = np.random.rand(512,2)
```

```python
x = data1[:,0]
y = data1[:,1]
plt.scatter(x,y)
```

```python
plt.scatter(data1[:, 0], data1[:, 1], c='g', marker='x')
plt.scatter(data2[:, 0], data2[:, 1], c='r', marker='o')
```

```python
n = 50
x = np.random.rand(n)
y = np.random.rand(n)
colors = np.random.rand(n)
area = (np.random.rand(n)*30)**2.5
plt.scatter(x,y, c=colors, s=area, alpha=0.5)
```

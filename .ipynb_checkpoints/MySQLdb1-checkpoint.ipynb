```python
import datetime
from datetime import datetime, date
rental = pd.read_csv('./mysqlcon/rental.csv')
rental['rent_mon'] = rental.rental_date.apply(lambda x:date(int(x.split('-')[0]),int(x.split('-')[1]),1 ).month)
rental.sort_values('rent_mon').groupby('rent_mon').size()
```

```python
sql = "select month(rental_date) as mon,count(*) as rental수 from rental group by mon order by mon; "
pd.read_sql(sql, sk)
```

```python
# payment 월별 전체 매출, 평균 매출, 결제횟수
import numpy as np
pay = pd.read_csv('./mysqlcon/payment.csv')
pay['month'] =  pay.payment_date.apply(lambda x: x[5:7])
#pd.DataFrame([pay.groupby('month').amount.sum(),pay.groupby('month').amount.mean(),pay.groupby('month').size()]).T
pay.groupby('month').agg({'payment_id':np.size,'amount':[np.mean, np.sum] })
```

```python
pd.DataFrame([pay.groupby('month').amount.sum(),pay.groupby('month').amount.mean(),pay.groupby('month').size()]).T
```

```python
sql = "select month(payment_date), avg(amount), sum(amount), count(*) from payment group by month(payment_date);"
pd.read_sql(sql, sk)
```

```python
# 구매횟수가 가장많은 고객(id), 총 구매액이 가장 큰 고객
pay = pd.read_csv('./mysqlcon/payment.csv')
cnt= pay.groupby('customer_id').agg({'payment_id':np.size}).sort_values('payment_id', ascending=False).head(1)
tot= pay.groupby('customer_id').agg({'amount':np.sum}).sort_values('amount',ascending=False).head(1)
pd.concat([cnt,tot], axis=1)
#idmax
#pay
```

```python
# 구매횟수가 가장많은 고객(id), 총 구매액이 가장 큰 고객
pay = pd.read_csv('./mysqlcon/payment.csv')
gbpay =  pay.groupby('customer_id').agg({'amount':np.sum, 'payment_id':np.size})
gbpay.loc[[gbpay.payment_id.idxmax(),gbpay.amount.idxmax()]]
#idmax
#pay
```

```python
# 구매횟수가 가장많은 고객(id), 총 구매액이 가장 큰 고객

sql = """
select * from
(select customer_id 구매수최다고객, count(payment_id) cnt from payment group by customer_id order by cnt desc limit 1) a,
(select customer_id 구매액최다고객, sum(amount) tot from payment group by customer_id order by tot desc limit 1) b;
"""

pd.read_sql(sql, sk)
```

```python
# 매출을 가장 많이 발생시킨 영화제목
#payment, firm
payment =pd.read_csv('./mysqlcon/payment.csv')[['rental_id', 'amount']]
film =pd.read_csv('./mysqlcon/film.csv')[['film_id', 'title']]
rental =pd.read_csv('./mysqlcon/rental.csv')[['rental_id', 'inventory_id']]
inventory =pd.read_csv('./mysqlcon/inventory.csv')[['inventory_id', 'film_id']]
rent_pay =  pd.merge(rental, payment, on='rental_id')
fi_inv =  pd.merge(film, inventory, on='film_id')
dffinal =  pd.merge(fi_inv, rent_pay, on='inventory_id')

print(dffinal.groupby('title').amount.sum().idxmax())
print( dffinal.groupby('title').amount.sum().max() )
```

```python
sql = """select f.title, sum(p.amount) 매출 
from payment p, inventory i, film f, rental r 
where r.rental_id = p.rental_id and f.film_id = i.film_id and r.inventory_id = i.inventory_id 
group by f.title
order by 매출 desc limit 3
"""
pd.read_sql(sql, sk)
```

```python
# store별 등급별 매출
store = pd.read_csv('./mysqlcon/store.csv')
payment = pd.read_csv('./mysqlcon/payment.csv')
inventory= pd.read_csv('./mysqlcon/inventory.csv')
rental = pd.read_csv('./mysqlcon/rental.csv')
film = pd.read_csv('./mysqlcon/film.csv')
rental.merge(inventory,on='inventory_id').merge(store, on='store_id').merge(payment, on='rental_id').merge(film, on='film_id').groupby(['store_id','rating']).amount.sum() 
```

```python
sql = """select s.store_id, sum(p.amount) tot,f.rating from store s, inventory i, payment p, rental r, film f
where r.inventory_id = i.inventory_id and i.store_id=s.store_id and r.rental_id = p.rental_id and i.film_id = f.film_id
group by f.rating, s.store_id order by tot; """
pd.read_sql(sql, sk)
```

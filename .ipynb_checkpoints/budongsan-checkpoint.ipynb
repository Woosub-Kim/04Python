```python
import pandas as pd
bd = pd.read_csv('budongsan.csv')[['매물번호', '연락처', '주소', '부동산', '보증금', '월세']]
bd[(bd.월세<=60)]
bd[(bd.월세<=60) & (bd.보증금<=1000)]
bd[(bd.보증금<=1000)]
bd[bd.보증금<=500].월세.min()
bd.loc[ bd[bd.보증금<=500].월세.idxmin() ]
```

```python
bd.보증금
```

```python
bd['업자']= bd.부동산.apply(lambda x: x.split('(')[1].replace(')', '') )
bd = bd.drop('연락처', axis=1)
```

```python
bd['부동산'] = bd.부동산.apply(lambda x: x.split('(')[0].replace('공인중개사', '') )
bd
```

```python
bd['주소'] = bd.주소.apply(lambda x: x.split()[2]+x.split()[3])
bd
```

```python
bd.sort_values('보증금').head()
```

```python
bd = bd[['매물번호', '주소', '부동산', '업자', '보증금', '월세']]
bd['예상전세'] = bd.보증금+bd.월세*100
bd
```

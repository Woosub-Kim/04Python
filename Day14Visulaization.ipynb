### Titanic

```python
df = pd.read_csv('titanic_data.csv')
df
```

```python
df = df[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin','Embarked']]
df.head()
```

```python
df.info()
```

```python
df[df.Age.isnull()]

cl3 = df[df.Pclass==3].Age.mean()
cl2 = df[df.Pclass==2].Age.mean()
cl1 = df[df.Pclass==1].Age.mean()

print(cl1, cl2, cl3)
```

```python
# age의 null값 처리
import math
age = []
for i in zip(df['Age'], df['Pclass']): 
    pc = i[1]
    a = i[0]
   
    if str(a) == 'nan':
        if pc ==1:
            age.append(int(cl1))
        elif pc==2:
            age.append(int(cl2))
        else:
            age.append(int(cl3))
    else:
        age.append(a)
df['Age'] = age
df.info()
```

```python
df.head(3)
```

```python
# df['who'] 컬럼만들기 13세 이하 Child, 초과 male/female
who = []
for i in zip(df.Sex, df.Age): # values사용 가능
    if i[1] <= 13:
        who.append('child')
    else:
        who.append(i[0])

df['who'] = who
df.head(10)
```

```python
# df['class'] third second first로 변경하기
def stringclass(cl):
    if cl == 1:
        return 'first'
    elif cl == 2:
        return 'second'
    else:
        return 'third'
df['class'] = df.Pclass.apply(stringclass)
df.head()
```

```python
# df['alive'] 컬럼 replace를 활용할 수 있다
df['alive'] = df.Survived.apply(lambda x: 'yes' if x==1 else 'no')
df.head()
```

```python
#d['alone'] 1 => yes 
df['alone'] = df.SibSp.apply(lambda x:'yes' if x==0 else 'no')
df.head()
```

```python
df.info()
```

```python
# cabin column 삭제
del df['Cabin']
df.head()
```

```python
# df의 who컬럼을 데이터 종류별로 bar차트
df.who.value_counts().plot(kind='bar')
plt.title('Who')
```

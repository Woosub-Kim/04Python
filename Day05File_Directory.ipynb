```python
def countBirths():
    ret = []
    for y in range(1880, 2016):
        count = 0
        filename = './names/yob{}.txt'.format(y)
        with open(filename, 'r') as f:
            data = f.readlines()
            for d in data:
                if d[-1] =='\n':
                    d=d[:-1]
                birth = d.split(',')[2]
                count+=int(birth)
            ret.append((y,count))
    return ret

#main
countBirths()
```

```python
# 연도별 총 여아, 총 남아
def countGenderBirths():
    ret = []
    for y in range(1880, 2016):
        gcount = 0
        bcount = 0
        count = 0
        filename = './names/yob{}.txt'.format(y)
        with open(filename, 'r') as f:
            data = f.readlines()
            for d in data:
                if d[-1] =='\n':
                    d=d[:-1]
                birth = d.split(',')[2]
                gender = d.split(',')[1] 
                if gender == 'M':
                    bcount+=int(birth)
                    count +=int(birth)
                else:
                    gcount += int(birth)
                    count +=int(birth)
            ret.append((y,gcount, bcount, count))
    return ret
with open('birth_by_years1.csv', 'w') as f:
    for year, gbirth, bbirth, total in countGenderBirths():
        data = '{},{},{},{}\n'.format(year, gbirth, bbirth, total)
        print(data)
        f.write(data)
with open('birth_by_years1.csv', 'r', encoding='cp949') as f:
    for i in f:
        print(i)    
```

```python
def popname(year):
    filename = './names/yob{}.txt'.format(year)
    with open(filename, 'r') as f:
        girl = []
        boy = []
        data = f.readlines()
        for d in data:
            if d[-1] =='\n':
                d=d[:-1]
                
            birth = int(d.split(',')[2])
            gender = d.split(',')[1]
            name = d.split(',')[0]
            #print(name, birth)
            if gender == 'M':
                count = -1
                flag = True
                if len(boy)==0:
                     boy.insert(0,(name,birth))
                else:
                    for i in boy: 
                        count += 1
                        if i[1] < birth:
                            #print('??')
                            boy.insert(count,(name,birth))
                            flag = False
                            break
                    if flag:
                        boy.append((name,birth))
                    if len(boy)>10:
                        boy.pop(-1)
            else:
                flag = True
                count = -1
                if len(girl)==0:
                     girl.insert(0,(name,birth))
                else:
                    for i in girl:
                        count += 1
                        if i[1] < birth:
                            
                            girl.insert(count,(name,birth))
                            flag = False
                            break
                    if flag:
                       
                        girl.append((name,birth))
                    if len(girl)>10:
                        girl.pop(-1)
            
    ret = [girl, boy]
    return ret
print(popname(2010))
```

```python
# 10만 이상의 숫자를 입력받은 후 천단위로 ','를 찍어 출력
#세자리씩 리스트에 담아 쉼표로 연결
def bignumber(n):
    L = []
    while True:
        if n<1000:
            L.insert(0,str(n))
            break
        else:
            L.insert(0,str(n%1000))
            n = n//1000
    for i in range(1,len(L)):
        if int(L[i])<10:
            L[i] = '00'+L[i]  
        elif int(L[i])<100:
            L[i] = '0'+L[i]
    return ','.join(L)
#문자열로 바꿔서 차곡차곡 쌓기
def bnumber(n):
    number = str(n)
    count = 0
    result = ''
    for i in number[::-1]:
        result = i+result
        if count ==2:
            result = ','+result
            count = 0
            continue
        count += 1
    if result[0]==',':
        return result[1:]
    return result

print(bnumber(1000))
print(bignumber(1000))
#입력받기
#number = int(input('정수 입력'))
#print(bignumber(number))
```

```python
# 특정 문서에서 특정 단어의 출현빈도 출력하는 함수
# 매개변수는 문서, 단어
def findword(doc, word):
    with open(doc, 'r') as f:
        result = {}
        data = f.read()
        data = data.lower()
        res = data.split()
        for i in res:
            if i[-1] in '.?!':
                i = i[:-1]
            result.setdefault(i,0)
            result[i] += 1
    return result.get(word)
findword('mydata.txt', 'of')
```

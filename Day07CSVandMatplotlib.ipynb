```python
def showGenderPop(dong):
    import csv
    with open('gender.csv') as f:
        data = csv.reader(f,delimiter=',')
        for row in data:
            if dong in row[0]:
                print(row[0])
                mresult = []
                fresult = []
                count = 0
                for i in row[3:104]:
                    mresult.append(int(i.replace(',','')))
                for i in row[106:]:
                    fresult.append(int(i.replace(',','')))
                break
    mpop = sum(mresult)
    fpop = sum(fresult)
    pop = [mpop, fpop]
    import matplotlib as mpl
    import matplotlib.pyplot as plt
    import matplotlib.font_manager as fm
    mpl.rcParams['axes.unicode_minus'] = False
    path = 'C:/windows/Fonts/malgun.ttf'
    font_name = fm.FontProperties(fname=path, size=50).get_name()
    plt.rc('font', family=font_name)
    plt.pie(pop, labels=['M', 'F'], colors=['blue', 'red'])
    plt.legend()
    plt.show()

showGenderPop('서울')    
```

```python
# 출근 시간대에 가장 승차인원이 많은 역은? 6-7, 7-8, 8-9시
# 중간과정: 출근시간대 승차인원을 dict로 정리
import csv
with open('subwaytime.csv') as f:
    data = csv.reader(f)
    next(data)
    next(data)
    station = {}
    #4부터 짝수 승차 홀수 하차
    for row in data:
        #print(row[3])
        #print(sum((map(int,row[8:13:2]))))
        station[row[3]]=sum((map(int,row[8:13:2])))
print(station)
m = max(list(station.values()))
#최대 승차인원?
print(m)
```

```python
# 출근 시간대에 가장 승차인원이 많은 역은? 6-7, 7-8, 8-9시
import csv
with open('subwaytime.csv') as f:
    data = csv.reader(f)
    next(data)
    next(data)
    station = ''
    num = 0
    #4부터 짝수 승차 홀수 하차
    for row in data:
        if num < sum((map(int,row[8:13:2]))):
            station = row[3]
            num = sum((map(int,row[8:13:2])))
print(station,num)
```

```python
# 특정 시간을 입력받은 후 주어진 시간에 승차인원이 가장 많은 역?
# 4시면 4, 5시면 6.....
def complex_station_ride(h):
    if h==3:
        raise Exception
    col = 2*h-4
    import csv
    with open('subwaytime.csv') as f:
        data = csv.reader(f)
        next(data)
        next(data)
        station = ''
        num = 0
        line = ''
        #4부터 짝수 승차 홀수 하차
        for row in data:
            if num < int(row[col]):
                station = row[3]
                num = int(row[col])
                line = row[1]
    #print(station,num)
    return (line,station,num,h,'ride')
def complex_station_out(h):
    col = 2*h-3
    import csv
    with open('subwaytime.csv') as f:
        data = csv.reader(f)
        next(data)
        next(data)
        station = ''
        num = 0
        line = ''
        #print(col)
        #4부터 짝수 승차 홀수 하차
        for row in data:
            if num < int(row[col]):
                station = row[3]
                num = int(row[col])
                line = row[1]
    #print(station,num)
    return (line,station,num,h,'out')
#main
print(complex_station_ride(4))
print(complex_station_ride(7))
print(complex_station_ride(19))
print(complex_station_ride(22))
print(complex_station_out(8))
print(complex_station_out(19))
print(complex_station_out(22))
```

```python
#페이지를 열어본 사람의 숫자..
with open('access_log', 'r') as f:
    iplist = []
    logs = f.readlines()
    for l in logs:
        if l.split()[-2]=='200' and l.split()[0] not in iplist:
            iplist.append(l.split()[0])
print(len(iplist))
print(iplist)
```

```python
#서버의 사용 용량의 합
with open('access_log', 'r') as f:
    usage = 0
    logs = f.readlines()
    for l in logs:
        if l.split()[-2]=='200':
            usage += int(logs[0].split()[-1]) 
print(usage)
with open('access_log', 'r') as f:
    usage = 0
    logs = f.readlines()
    for l in logs:
        if l.split()[-1].isdigit():
            usage += int(l.split()[-1]) 

print(usage)
```

```python
# ip별 사용량의 합
with open('access_log', 'r') as f:
    ip_usage = {} 
    logs = f.readlines()
    for l in logs:
        if l.split()[-1].isdigit():
            ip_usage.setdefault(l.split()[0],0)
            ip_usage[l.split()[0]]+=int(l.split()[-1]) 
print(sorted(ip_usage, key=lambda x:x[1], reverse=True))
print(ip_usage)
print(sum(ip_usage.values()))
len(ip_usage)

```

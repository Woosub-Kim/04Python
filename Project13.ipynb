```python
import pandas as pd
import numpy as np
x = np.load('./project/tatanic_x_train.npy')
y = np.load('./project/tatanic_y_train.npy')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3)
```

```python
pd.DataFrame(y_train).iloc[:, 8:18]
```

```python
pd.DataFrame(x_train).iloc[:, 8:18]
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(criterion='gini', n_estimators=25)
forest.fit(x_train, y_train)
pr = forest.predict(x_test)
from sklearn.metrics import classification_report
print( classification_report(y_test,pr ) )
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
param_range = {'C': [0.001,0.01,0.1,1,10,100,1000]}
grid = GridSearchCV(
    estimator=LogisticRegression(), 
    param_grid=param_range,
    scoring='accuracy',
    cv=10,
    n_jobs=1,
)
grid1 = grid.fit(x_train, y_train)
grid1.best_params_
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression(C=1)
log.fit(x_train, y_train)
print(classification_report(y_test, log.predict(x_test) ) )
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
param_range = {'C': [0.01,0.1,1,10,100], 'gamma': [100,10,1,0.1,0.01]}
grid = GridSearchCV(
    estimator=SVC(), 
    param_grid=param_range,
    scoring='accuracy',
    cv=10,
    n_jobs=1,
)
grid1 = grid.fit(x_train, y_train)
grid1.best_params_
```

```python
print( classification_report( y_test, SVC(C=1, gamma=1).fit(x_train, y_train).predict(x_test)))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier
log = LogisticRegression()
tree = DecisionTreeClassifier()
gauss = GaussianNB()
hard_voter = VotingClassifier(
    estimators=[('lr', log),('dt', tree)],
    voting='hard'
)
soft_voter = VotingClassifier(
    estimators=[('lr', log),('dt', tree)],
    voting='soft'
)
models = [log, tree, gauss, hard_voter, soft_voter]
for model in models:
    model.fit(x_train, y_train)
    score = model.score(x_test, y_test)
    print(model)
    print(score)
    print('+'*30)
```

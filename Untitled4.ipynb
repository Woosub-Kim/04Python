```python
import pandas as pd
crime = pd.read_csv('Crime.csv', encoding = 'cp949')
crime
```

```python
crime.columns
```

```python
col =  list(crime.iloc[:,1])
col
```

```python
crime_df = crime.iloc[:,2:-2].T
crime_df.columns = col
crime_df.head(20)
```

```python
crime_df.dropna()
```

```python
crime_city = []
for i in crime_df.index:
    if i=='서울':
        i = '서울특별시'
    elif i =='세종':
        i= '세종특별자치시'
    elif len(i) == 2:
        i = i+ '광역시'
    else:
        i = i.split()[1]+'시'
    crime_city.append(i)
crime_city
        
```

```python
crime_df['도시명'] = crime_city
crime_df
```

```python
pop = pd.read_csv('pop.csv', encoding='cp949', header=2,).iloc[:,4:].T
pop
```

```python
cityName = pop.index
population = {}
for i in cityName:
    population[i]=int(pop.loc[i])
population.pop('세종시')
population
```

```python
city =  list(pop.index)
cityName = []
for j in city:
    if j[-1]=='시':
        cityName.append(j)


print(len(cityName))
print(len(crime_df.index))


for i in range(86):
    print(crime_df.index[i], cityName[i])
```

```python
crime_df['인구'] = crime_df.도시명.map(population)
crime_df
```

```python
crime_item =  list(crime_df.columns)[:-2]
for i in crime_item:
    crime_df[i] = crime_df[i]/crime_df.인구
crime_df
```

```python
crime_df['성범죄'] =  crime_df.강간 + crime_df.유사강간 + crime_df.강제추행 + crime_df.기타강간강제추행등
crime_df
```

```python
crime_df.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
x = crime_df[['살인기수', '살인미수등', '강도', '방화','폭행', '체포감금', '협박',\
             '약취유인', '폭력행위등', '공갈','공갈', '손괴', '직무유기', '직권남용',
       '증수뢰', '통화', '문서인장', '유가증권인지', '사기', '횡령', '배임', '성풍속범죄', '도박범죄',
       '특별경제범죄', '마약범죄', '보건범죄', '환경범죄', '교통범죄', '노동범죄', '안보범죄', '선거범죄',
       '병역범죄', '기타범죄']]
y = crime_df.성범죄
x_std = StandardScaler().fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x_std, y, test_size = 0.3)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
pred =  lr.predict(x_test)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, pred)
print(report)
```

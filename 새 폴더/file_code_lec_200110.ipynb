```python
# '비둘기.jpg' 파일을 복사 저장 '비둘기_복사.jpg'
# 이미지, 영상, 소리, 코드 => binary file => 'rb', 'wb'
```

```python
f = open('비둘기.jpg', 'rb')
g = open('비둘기_복사1.jpg', 'wb')

bufsize=1024

img = f.read(bufsize)

while img:
    g.write(img)    
    img = f.read(bufsize)

f.close()
g.close()
```

### merge

```python
customers = {'CustomerID':[10,11], 'Name': ['Mike', 'Marcia'], 'Address': ['Mike home', 'Marcia home']}
customers =  pd.DataFrame(customers)
customers
```

```python
orders = {'CustomerID':[10,11,10], 'OrderDate': [datetime(2014,12,1),datetime(2014,12,1),datetime(2014,12,1)]}
orders =  pd.DataFrame(orders)
orders
```

```python
pd.merge(customers,orders)
# pd.merge(customers,orders, on='CustomerID', how='inner') default는 inner, 공통컬럼을 자동으로 찾아줌
```

```python
customers.merge(orders)
```

```python
pd.merge(customers, orders, on='CustomerID', how='right')
```

```python
pd.merge(customers, orders, on='CustomerID', how='outer')
```

```python
left_data = {'key1':['a','b', 'c'], 'key2':['x','y', 'z'], 'lval':[0,1,2]}
right_data =  {'key1':['a','b', 'c'], 'key2':['x','a', 'z'], 'rval':[6,7,8]}

left=pd.DataFrame(left_data)
right = pd.DataFrame(right_data)

print(left)
print(right)
```

```python
pd.merge(left, right)
```

```python
pd.merge(left, right, how='outer')
```

```python
pd.merge(left, right, on='key1')
```

```python
pd.merge(left, right, on='key2')
```

```python
pd.merge(left, right, on='key2', how='left')
```

```python
pd.merge(left, right, on='key2', how='right')
```

```python
pd.merge(left, right, on='key2', how='outer')
```

```python
pd.merge(left, right, left_index=True, right_index=True)
```

```python
other_data =  {'key1':['a','b', 'c'], 'key2':['x','y', 'z'], 'oval':[12,23,45]}
other = pd.DataFrame(other_data)
other
```

```python
left.merge(right, how='outer').merge(other, how='outer')
```

```python
sensor_readings = pd.read_csv('./Learning-Pandas-Second-Edition-master/data/sensors.csv')
sensor_readings
```

```python
user1 = sensor_readings.copy()
user1['who'] = 'Mike'
user2 = sensor_readings.copy()
user2['reading'] = user2.reading*2
user2['who'] = 'Mikael'
print(user1)
print(user2)
```

```python
multi_user_sensor_data = pd.concat([user1, user2]).set_index(['who', 'interval', 'axis'])
multi_user_sensor_data
```

```python
multi = multi_user_sensor_data.xs(1, level='interval') #interval =1인 것만 가져옴
multi
```

```python
multi = multi.drop_duplicates()
multi
```

```python
multi.unstack()
```

```python
multi.unstack(['who', 'axis'])
```

```python
multi.unstack(level=1)
```

```python
multi.unstack(level=0)
```

```python
multi.unstack(level=2)
```

### melting

```python
data = pd.DataFrame({'Name':['Mike', 'Mikael'], 'Height':[6.1,6.0], 'Weight':[220,185]})
data
```

```python
pd.melt(data, id_vars=['Name'], value_vars=['Height', 'Weight'])
```

```python
stkd1 = data.stack()
stkd1
```

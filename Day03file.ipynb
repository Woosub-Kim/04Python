```python
#input함수로 입력을 받아서 파일에 기록하고 파일 전체 내용을 출력하는 함수
def writeAndRead():
    message = input('파일에 입력할 메세지')
    f = open('text.txt', 'w')
    msg = message.split('>')
    for i in range(len(msg)):
        msg[i] += '\n'
    f.writelines(msg)
    f.close()
    f = open('text.txt')
    text = f.read()
    print(text)
    f.close()
# main
writeAndRead()
        
```

```python
#stockcode.txt파일을 stockcode_copy.txt파일로 복사해 저장하기
f = open('stockcode.txt', 'r')
codes = f.readlines()
f.close()
g = open('stockcode_copy.txt', 'w')
g.writelines(codes)
g.close()
```

```python
#for문으로 파일 찍어내기
for i in range(11):
    title = "text{}.txt".format(i)
    f=open(title, 'w')
    f.write('hi\n'*i)
    f.close()
```

```python
#googoo.jpg를 비둘기.jpg파일로 복사저장
#이미지 영상, 소리 코드는 binary file => 'rb', 'wb', 사용 
f = open('googoo.jpg', 'rb')
g = open('비둘기1.jpg', 'wb')
#g.write(f.read())
#데이터절약을 위해
img = f.read(1024)
while img:
    g.write(img)
    img=f.read(1024)
f.close()
g.close()

```

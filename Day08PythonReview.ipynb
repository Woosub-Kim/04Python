### Set

```python
# 원소가 없는 set을 만드세요
s = set()
print(s)

# 원소 'sky'와 'earth'를 추가하세요
s.add('sky')
s.add('earth')
print(s)

# (1,2) 두 원소를 추가하세요
s.update((1,2))
print(s)

# 'sky'를 삭재하세요
#s.remove('sky')
s.discard('sky')
print(s)

# 다음의 집합에서
A = {1,2,3,4,5,6}
B = {1,2,3,7,8,9}
C ={1,2,3}

# A,B 두 집합의 합집합을 출력
print(A.union(B))
print(A|B)

# A의 B에 대한 집합의 차집합을 출력
print(A.difference(B))
print(A-B)

# A,B 두 집합의 교집합을 구하세요
print(A.intersection(B))
print(A&B)

# A,B의 대칭차집합
print(A.symmetric_difference(B))
print(A^B)

# C가 A의 부분집합인지 확인
print(C.issubset(A))
print(C <= A)
#print({1,2,3}<={1,2,3}) True
#print({1,2,3}<{1,2,3}) False
#print({1,2}<={1,2,3}) True
#print({1,2}<{1,2,3}) True

# A가 C를 포함하는지 확인
print(A.issuperset(C))
print(A >= C)

# B의 길이를 출격
print(len(B))

# B의 최댓값 출력
print(max(B))

# B의 최솟값 출력
print(min(B))

# B의 합을 출력
print(sum(B))

# B의 평균을 출력
print(sum(B)/len(B))

# B를 내림차순으로 정렬
print(sorted(B, reverse=True))

# B에 정수 5가 포함되어 있는지 확인하기
print(5 in B)
```

### Dictionary

```python
# 'mydict'라는 dict를 만드시오
mydict = {}

# 'red':1, 'blue':3, 'green':4 세 요소를 추가하시오
mydict['red']=4
mydict['blue']=3
mydict['green']=1
print(mydict)

# key에 'blue'기 있는지 확인하시오
print('blue' in mydict)

# mydict의 길이를 출력하시오
print(len(mydict))

# 'red'의 value를 출력하시오
print(mydict['red'])

# 'red'의 value를 10으로 변경하시오
mydict['red'] = 10
print(mydict)

# key값만 출력하시오
print(mydict.keys())

# value값만 출력하시오
print(mydict.values())

# key와 value를 tuple로 출력하시오
print(mydict.items())

# key기준으로 정렬하시오
print(sorted(mydict.items(), key = lambda x: x[0]))

# value기준으로 정렬하시오
print(sorted(mydict.items(), key = lambda x: x[1]))

# key 'blue'를 삭제하시오
#del mydict['blue']
mydict.pop('blue')
print(mydict)

# key 'blue'의 값을 오류없이 출력하시오
print(mydict.get('blue'))

# mydict에 key 'pink'값을 출력하되 없으면 20추가
mydict.setdefault('pink', 20)
print(mydict)

# value에 20이 있는지 확인
print(20 in mydict.values())

# 'yellow':30을 추가하시오
mydict['yellow'] = 30

# yourdict = {'파랑':50, '빨강':70}을 합하시오
yourdict = {'파랑':50, '빨강':70}
mydict.update(yourdict)
print(mydict)
```

```python
# text = 'python is fun' 중에서 문자를 대문자로 변환해서 출력하세요
text = 'python is fun'
for c in text:
    print(c.upper(), end=' ')
print()

# 1~20까지 출력하세요
j = 1
while j<=20:
    print(j, end=' ')
    j +=1
print()

# 1~20까지 짝수만 리스트에 담으세요 for, while, list comprehension
L = []
j = 0
while j<20:
    j += 1
    if j%2!=0:
        continue
    L.append(j)
print(L)

L = []
for i in range(1,21):
    if i%2==0:
        L.append(i)
print(L)

print([i for i in range(1,21) if i%2==0])
# for문과 while문을 활용해서 1~10까지 구하세요

#for문 사용
count1 = 0
for i in range(1,11):
    count1 += i
print(count1)

#while문 사용
j=1
count2 = 0
while j<=10:
    count2 += j
    j+=1
print(count2)

# 1~10사이의 정수중에서 짝수만 리스트에 담으세요 for문으로
L = []
for i in range(1,11):
    if i%2==0:
        L.append(i)
print(L)

# 위 문제를 ListComprehension으로 구현하시오
List = [i for i in range(1,11) if i%2==0]
print(List)

# computational thinking에서 모음만 리스트에 담으세요
L =[]
for c in 'computational thinking':
    if c in 'aeiou':
        L.append(c)
print(L)

## 점수를 5회 입력받은 후에 평균을 출력하시오
# score = []
# while True:
#     if len(score)==5:
#         break
#     try:
#         s = int(input('점수를 입력하시오'))
#         score.append(s)
#
#     except:
#         print('정수를 입력하세요')
# import numpy as np
# print(np.mean(score))

# text = 'hello world'에서 각 문자를 인덱스와 함께 tuple로 출력하시오
text = 'hello world'
for i,c in enumerate(text):
    print((i,c), end='')
print()

#1~50까지의 수에서 짝수함과 홀수합으로 각각 출력하세요
jak=0
hol=0
k=1
while k<=50:
    if k%2==0:
        jak+=k
    else:
        hol+=k
    k+=1
print('짝수 합: {} \n홀수 합: {}'.format(jak,hol))
```

```python
# mytest.txt파일을 열어서 아래의 텍스트를 작성하세요
"""
'hello python'
'python progamming'
'python world'
'bigdata'
'artificial intelligence'
"""
with open('mytest.txt', 'w') as f:
    f.write('hello python\n')
    f.write('python progamming\n')
    f.write('python world\n')
    f.write('bigdata\n')
    f.write('artificial intelligence\n')
# mytest.txt.파일을 4가지 방법으로 출력하시오
with open('mytest.txt', 'r') as f:
    data = f.read()
    print(data)
with open('mytest.txt', 'r') as f:
    data = f.readline()
    while data:
        print(data)
        data = f.readline()
with open('mytest.txt', 'r') as f:
    data = f.readlines()
    print(data)
with open('mytest.txt', 'r') as f:
    for l in f:
        print(l)


# mytest파일을 열어서 아래와 같이 출력하세요
"""
1,'hello python'
2,'python progamming'
3,'python world'
4,'bigdata'
5,'artificial intelligence'
"""
f = open('mytest.txt', 'r')
for i,l in enumerate(f):
    print('{},'.format(i+1),l)
f.close()

# 'mytest.txt' 파일을 with구문을 활용해서 읽어서 출력하시오
with open('mytest.txt', 'r') as f:
    data = f.read()
    print(data)

# test_list = ['a', 'b', 'c'] pickle파일로 저장하세요
test_list = ['a', 'b', 'c'] 
import pickle
with open('mytest.pickle', 'wb') as f:
    pickle.dump(test_list, f)

# 위에서 저장한 피클파일을 읽어 출력하세요    
with open('mytest.pickle', 'rb') as f:
    data = pickle.load(f)
    print(data)
    
```

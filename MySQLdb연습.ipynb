```python
import MySQLdb
import pandas as pd
```

```python
sk= MySQLdb.connect(user='root', db='sakila', password='1234', port=3306,charset='utf8' )
df = pd.read_sql('select * from customer', sk)
df.head()
```

```python
tl = pd.read_sql('show tables;',sk)
tl
```

```python
for i in range(len(tl)):
    df=pd.read_sql('select * from {};'.format(tl.iloc[i].values[0]), sk)

    df.to_csv('./csvfiles/sakila/{}.csv'.format(tl.iloc[i].values[0]))
```

```python
for i in tl.values:
    df = pd.read_sql('select * from {};'.format(i[0]), sk)
    df.to_csv('./csvfiles/sakila/{}2.csv'.format(i[0]))
```

### sql

```python
import sqlite3
```

```python
msft = pd.read_csv('./Learning-Pandas-Second-Edition-master/data/msft.csv')
msft['Symbol'] = 'MSFT'
aapl = pd.read_csv('./Learning-Pandas-Second-Edition-master/data/aapl.csv')
aapl['Symbol'] = 'AAPL'
```

```python
connection =  sqlite3.connect('Learning-Pandas-Second-Edition-master/data/stock.sqlite')
msft.to_sql('STOCK_DATA', connection, if_exists='replace')
aapl.to_sql('STOCK_DATA', connection, if_exists='append')
connection.commit()
connection.close()
```

```python
connection = sqlite3.connect('Learning-Pandas-Second-Edition-master/data/stock.sqlite')
stocks = pd.io.sql.read_sql('select * from stock_data;', connection)
stocks.head()
```

```python
connection = sqlite3.connect('Learning-Pandas-Second-Edition-master/data/stock.sqlite')
query = "SELECT * FROM stock_data WHERE volume>29000000 and symbol = 'MSFT';"
items = pd.io.sql.read_sql(query, connection)
items.head()
```
